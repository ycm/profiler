```python
import json
import numpy as np
import pandas as pd
import pickle as pkl
import matplotlib.pyplot as plt

from scipy.stats import mode
```

```python
with open('../data/moby/jsons/item_to_form.json') as f:
    item_to_form = json.load(f)
```

```python
df = pd.read_csv('stats/RADIAL.tsv', sep='\t')
item_to_radial_acc_full_dataset = dict(zip(df.Item.astype(str), df.AccOnFullDataset))
```

```python
with open('../data/moby/jsons/form_to_items.json') as f:
    form_to_items = json.load(f)
```

```python
with open('stats/INTERCEPT_NO_SEP.json') as f:
    item_to_intercept_acc_no_sep_full_dataset = json.load(f)
```

```python
df = pd.read_csv('stats/INTERCEPT_SEP.tsv', sep='\t')
item_to_intercept_acc_with_sep_full_dataset = dict(zip(df.Item.astype(str), df.AccOnFullDataset))
```

```python
df = pd.read_csv('stats/INTERCEPT_SEP_BY_ANNOTE.tsv', sep='\t')
item_to_intercept_acc_with_sep_full_dataset_BY_ANNOTE = dict(zip(df.Item.astype(str), df.AccOnFullDataset))
```

```python
with open('stats/INTERCEPT_SEP_100FOLD_accs.json') as f:
    item_to_intercept_accs_with_sep_100folds = json.load(f)
```

```python
with open('stats/INTERCEPT_SEP_100FOLD_accs_BY_ANNOTE.json') as f:
    item_to_intercept_accs_with_sep_100folds_BY_ANNOTE = json.load(f)
```

```python
fig, axs = plt.subplots(12, 3)
fig.set_figwidth(16)
fig.set_figheight(48)
fig.set_dpi(306)

for grade_idx, grade in enumerate(('Grade1', 'Grade2', 'Grade3', 'Grade4')):
    for term_idx, term in enumerate(('Fall', 'Winter', 'Spring')):
        form = grade + term
        for item in form_to_items[form]:
            subplot_x = 3 * grade_idx + term_idx
            subplot_y = form_to_items[form].index(item)
            
            axs[subplot_x, subplot_y].set_title(' '.join((form, item)))
            axs[subplot_x, subplot_y].set_xlim(0.4, 1)
            axs[subplot_x, subplot_y].set_ylim(0, 40)
            
            axs[subplot_x, subplot_y].hist(
                item_to_intercept_accs_with_sep_100folds[item],
                alpha=0.5,
                color='coral',
                label='intercept, with sep, 100-folds'
            )
            
            axs[subplot_x, subplot_y].hist(
                item_to_intercept_accs_with_sep_100folds_BY_ANNOTE[item],
                alpha=0.5,
                color='mediumseagreen',
                label='^by annotation'
            )
            
            
            # radial
            axs[subplot_x, subplot_y].axvline(
                item_to_radial_acc_full_dataset[item],
                color='red',
                label='radial, full data'
            )
            
            # naive, fulldataset, nosep line:
            axs[subplot_x, subplot_y].axvline(
                item_to_intercept_acc_no_sep_full_dataset[item] / 100,
                color='orange',
                label='intercept, full data, no sep'
            )
            
            # withsep, fulldata line
            axs[subplot_x, subplot_y].axvline(
                item_to_intercept_acc_with_sep_full_dataset[item],
                color='blue',
                label='intercept, full data, with sep'
            )
            # by annote
            axs[subplot_x, subplot_y].axvline(
                item_to_intercept_acc_with_sep_full_dataset_BY_ANNOTE[item],
                color='blue',
                linestyle='--',
                label='^by annotation'
            )
            
            # withsep, mean 100fold line
            axs[subplot_x, subplot_y].axvline(
                np.mean(item_to_intercept_accs_with_sep_100folds[item]),
                color='purple',
                label='intercept, with sep, 100-fold median'
            )
            # by annote
            axs[subplot_x, subplot_y].axvline(
                np.mean(item_to_intercept_accs_with_sep_100folds_BY_ANNOTE[item]),
                color='purple',
                linestyle='--',
                label='^by annotation'
            )
axs[0,0].legend()
plt.tight_layout()
plt.savefig('stats/big-hist.png')
```

```python
items = []

avgannote_min = []
avgannote_max = []
avgannote_mean = []
avgannote_median = []
avgannote_mode = []
avgannote_mode_occurrences = []

single_min = []
single_max = []
single_mean = []
single_median = []
single_mode = []
single_mode_occurrences = []

for item, accs in item_to_intercept_accs_with_sep_100folds.items():
    items.append(item)
    
    avgannote_min.append(min(accs))
    avgannote_max.append(max(accs))
    avgannote_mean.append(np.mean(accs))
    avgannote_median.append(np.median(accs))
    avgannote_mode.append(mode(accs)[0][0])
    avgannote_mode_occurrences.append(mode(accs)[1][0])
    
    accs_by_annote = item_to_intercept_accs_with_sep_100folds_BY_ANNOTE[item]
    single_min.append(min(accs_by_annote))
    single_max.append(max(accs_by_annote))
    single_mean.append(np.mean(accs_by_annote))
    single_median.append(np.median(accs_by_annote))
    single_mode.append(mode(accs_by_annote)[0][0])
    single_mode_occurrences.append(mode(accs_by_annote)[1][0])
```

```python
pd.DataFrame({
    'item': items,
    'form': [item_to_form[i] for i in items],
    'avgannote_min': avgannote_min,
    'avgannote_max': avgannote_max,
    'avgannote_mean': avgannote_mean,
    'avgannote_median': avgannote_median,
    'avgannote_mode': avgannote_mode,
    'avgannote_mode_occurrences': avgannote_mode_occurrences,
    'single_min': single_min,
    'single_max': single_max,
    'single_mean': single_mean,
    'single_median': single_median,
    'single_mode': single_median,
    'single_mode_occurrences': single_mode_occurrences,
}).to_csv('stats/hist_stats.csv', index=0)
```
