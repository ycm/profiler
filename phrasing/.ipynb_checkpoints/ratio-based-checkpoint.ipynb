```python
PERF_MAT_DIR = '../output/performance-matrix/all-fixed/'
```

```python
def get_sliding_windows(lst, window_size=5):
    return [tuple(lst[i:i + window_size]) for i in range(len(lst)) if i < len(lst) - window_size + 1]
```

```python
item_to_gold_readings = defaultdict(list)
for item in items:
    gold_reading_dir = os.path.join(PERF_MAT_DIR, item + '-182')
    for gold_reading_table in os.listdir(gold_reading_dir):
        if not gold_reading_table.endswith('tsv'):
            continue
        item_to_gold_readings[item].append(
            pd.read_csv(os.path.join(gold_reading_dir, gold_reading_table), sep='\t'))
```

```python
def compute_diff_score(lst1, lst2):
    return np.median(np.abs(np.array(lst1) - np.array(lst2)))
```

```python
def color_to_val(color):
    if color == 'green':
        return 3
    if color == 'gold':
        return 2
    return 1
```

```python
def to_color_2(val):
    if val == 3:
        return 'green'
    if val == 2:
        return 'gold'
    if val == 1:
        return 'red'
    assert 0
```

```python
def to_color(val):
    if PER_ANNOTE:
        return val
    if val == 3:
        return 'green'
    if val <= 2:
        return 'red'
    return 'gold'
```

```python
rows = []

for item, gold_readings in item_to_gold_readings.items():
    
    perf_mat_height = len(gold_readings[0])
    token_indices = [x for x in range(perf_mat_height) if x % 2 != 0]
    
    for window in get_sliding_windows(token_indices):
        _token_idxs = ','.join([str((x - 1) // 2) for x in window])
        _tokens = ','.join([gold_readings[0].token[x] for x in window])
        
        
        ratio_sets = []
        for gold_reading in gold_readings:
            pause_lengths_in_window = [gold_reading.nframes[x] for x in window]

            if pause_lengths_in_window[1]:
                ratio_0_over_1 = pause_lengths_in_window[0] / pause_lengths_in_window[1]
            else:
                ratio_0_over_1 = 0
            if pause_lengths_in_window[2]:
                ratio_0_over_2 = pause_lengths_in_window[0] / pause_lengths_in_window[2]
            else:
                ratio_0_over_2 = 0
            if pause_lengths_in_window[3]:
                ratio_0_over_3 = pause_lengths_in_window[0] / pause_lengths_in_window[3]
            else:
                ratio_0_over_3 = 0
            if pause_lengths_in_window[4]:
                ratio_0_over_4 = pause_lengths_in_window[0] / pause_lengths_in_window[4]
            else:
                ratio_0_over_4 = 0
            if pause_lengths_in_window[2]:
                ratio_1_over_2 = pause_lengths_in_window[1] / pause_lengths_in_window[2]
            else:
                ratio_1_over_2 = 0
            if pause_lengths_in_window[3]:
                ratio_1_over_3 = pause_lengths_in_window[1] / pause_lengths_in_window[3]
            else:
                ratio_1_over_3 = 0
            if pause_lengths_in_window[4]:
                ratio_1_over_4 = pause_lengths_in_window[1] / pause_lengths_in_window[4]
            else:
                ratio_1_over_4 = 0
            if pause_lengths_in_window[3]:
                ratio_2_over_3 = pause_lengths_in_window[2] / pause_lengths_in_window[3]
            else:
                ratio_2_over_3 = 0
            if pause_lengths_in_window[4]:
                ratio_2_over_4 = pause_lengths_in_window[2] / pause_lengths_in_window[4]
            else:
                ratio_2_over_4 = 0
            if pause_lengths_in_window[4]:
                ratio_3_over_4 = pause_lengths_in_window[3] / pause_lengths_in_window[4]
            else:
                ratio_3_over_4 = 0
            
            ratios = [ratio_0_over_1,ratio_0_over_2,ratio_0_over_3,ratio_0_over_4,\
                      ratio_1_over_2,ratio_1_over_3,ratio_1_over_4,ratio_2_over_3,\
                      ratio_2_over_4,ratio_3_over_4]
            _ratios_str = ','.join(['{:.4f}'.format(r) for r in ratios])
            ratio_sets.append(_ratios_str)
        _ratio_sets_str = ';'.join(ratio_sets)
        
        row = {
            'item': item,
            'WINDOW_SIZE': 5,
            'WINDOW_MIN_CORRECT': 3,
            'NUM_TOKENS_IN_PASSAGE': len(token_indices),
            'token_indices_in_window': _token_idxs,
            'tokens_in_window': _tokens,
            'ratio_sets_in_window': _ratio_sets_str,
            
        }
        
        rows.append(row)
```
