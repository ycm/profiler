```python
import json
import statistics
import pandas as pd
import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
df = pd.read_csv('data/all_sessions_only_session_scores.csv')
with open('../data/moby/jsons/session_to_form.json') as f:
    session_to_form = json.load(f)
with open('../data/moby/jsons/form_to_sessions.json') as f:
    form_to_sessions = json.load(f)
with open('../data/moby/jsons/session_to_acc.json') as f:
    session_to_acc = json.load(f)
with open('../data/moby/jsons/session_to_wcpm.json') as f:
    session_to_wcpm = json.load(f)
```

```python
form_to_wcpms = {
    form: [session_to_wcpm[str(session)] for session in sessions]
    for form, sessions in form_to_sessions.items()
}

form_to_accs = {
    form: [session_to_acc[str(session)] for session in sessions]
    for form, sessions in form_to_sessions.items()
}
```

```python
for form, wcpms in form_to_wcpms.items():
    plt.title(form + ' WCPMs distribution')
    plt.hist(wcpms, edgecolor='black', bins=10)
    plt.ylim((0, 30))
    plt.xlim((0, 200))
    plt.show()
```

```python
for form, accs in form_to_accs.items():
    plt.title(form + ' Accs distribution')
    plt.hist(accs, edgecolor='black', bins=10)
    plt.ylim((0, 85))
    plt.xlim((0.5, 1))
    plt.show()
```

```python
session_to_wcpm_zscore = {}
for session, wcpm in session_to_wcpm.items():
    form = session_to_form[session]
    wcpms = form_to_wcpms[form]
    
    x = wcpm
    mu = statistics.mean(wcpms)
    sigma = statistics.stdev(wcpms)
    
    zscore = (x - mu) / sigma
    
    session_to_wcpm_zscore[session] = zscore
```

```python
session_to_acc_zscore = {}
for session, acc in session_to_acc.items():
    form = session_to_form[session]
    accs = form_to_accs[form]
    
    x = acc
    mu = statistics.mean(accs)
    sigma = statistics.stdev(accs)
    
    zscore = (x - mu) / sigma
    
    session_to_acc_zscore[session] = zscore
```

```python
form_to_grade = {
    'Grade1Fall':   'Grade1',
    'Grade1Spring': 'Grade1',
    'Grade1Winter': 'Grade1',
    'Grade2Fall':   'Grade2',
    'Grade2Spring': 'Grade2',
    'Grade2Winter': 'Grade2',
    'Grade3Fall':   'Grade3',
    'Grade3Spring': 'Grade3',
    'Grade3Winter': 'Grade3',
    'Grade4Fall':   'Grade4',
    'Grade4Spring': 'Grade4',
    'Grade4Winter': 'Grade4',
}
```

```python
grade_to_wcpms = defaultdict(list)
for form, wcpms in form_to_wcpms.items():
    grade = form_to_grade[form]
    for wcpm in wcpms:
        grade_to_wcpms[grade].append(wcpm)

grade_to_accs = defaultdict(list)
for form, accs in form_to_accs.items():
    grade = form_to_grade[form]
    for acc in accs:
        grade_to_accs[grade].append(acc)
```

```python
for grade, wcpms in grade_to_wcpms.items():
    plt.title(grade + ' WCPMs distribution')
    plt.hist(wcpms, edgecolor='black', bins=10)
    plt.ylim((0, 100))
    plt.xlim((0, 200))
    plt.show()
```

```python
for grade, accs in grade_to_accs.items():
    plt.title(grade + ' Accs distribution')
    plt.hist(accs, edgecolor='black', bins=20)
    plt.ylim((0, 200))
    plt.xlim((.6, 1))
    plt.show()
```

```python
session_to_wcpm_zscore_by_grade = {}
for session, wcpm in session_to_wcpm.items():
    form = session_to_form[session]
    grade = form_to_grade[form]
    wcpms = grade_to_wcpms[grade]
    
    x = wcpm
    mu = statistics.mean(wcpms)
    sigma = statistics.stdev(wcpms)
    
    zscore = (x - mu) / sigma
    
    session_to_wcpm_zscore_by_grade[session] = zscore
```

```python
session_to_acc_zscore_by_grade = {}
for session, acc in session_to_acc.items():
    form = session_to_form[session]
    grade = form_to_grade[form]
    accs = grade_to_accs[grade]
    
    x = acc
    mu = statistics.mean(accs)
    sigma = statistics.stdev(accs)
    
    zscore = (x - mu) / sigma
    
    session_to_acc_zscore_by_grade[session] = zscore
```
